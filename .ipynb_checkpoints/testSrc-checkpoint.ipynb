```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("data.csv", encoding='cp949')
df.head(3)
```

```python
# 득점 분석 꺾은선 그래프

teamName = '맨시티'
name = df[df['홈'] == teamName]
teamScore = name['득점']
time = name['시간']
x = time
y = teamScore
plt.plot(x, y)
plt.show()
```

```python
# 실점 분석 꺾은선 그래프

teamName = '맨시티'
nameDf = df[df['홈'] == teamName]

time = nameDf['시간']
x = time

teamLoss = nameDf['실점']
y = teamLoss

plt.plot(x, y)
plt.show()
```
