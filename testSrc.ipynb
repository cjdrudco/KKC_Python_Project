```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("data.csv", encoding='cp949')
df.head(3)
```

```python
# 득점 분석 꺾은선 그래프

teamName = '맨시티'
name = df[df['홈'] == teamName]
teamScore = name['득점']
time = name['시간']
x = time
y = teamScore
plt.plot(x, y)
plt.show()
```

```python
# 실점 분석 꺾은선 그래프

teamName = '맨시티'
nameDf = df[df['홈'] == teamName]

time = nameDf['시간']
x = time

teamLoss = nameDf['실점']
y = teamLoss

plt.plot(x, y)
plt.show()
```

```python
# 유효슈팅 당 득점 분석 막대그래프

teamName = '맨시티'
nameDf = df[df['홈'] == teamName]


teamShoot = nameDf['유효슈팅']
teamScore = nameDf['득점']

x = teamShoot
y = teamScore

plt.bar(x, y)
plt.show()
```

```python
# 점유율 당 득점 분석 산점도 

teamName= '맨시티'
teamDf = df[df['홈'] == teamName]

share = teamDf['볼점유율']
score = teamDf['득점']

x = share
y = score

plt.scatter(x, y)
plt.show()
```

```python
# 최근 3경기 전적 표

teamName = '맨시티'

teamAway = df[df['원정'] == teamName]
teamHome = df[df['홈'] == teamName]

final = pd.concat([teamAway, teamHome])
res = final.sort_values(by='시간')

res.head(3)
```

```python
#홈팀과 원정팀의 시간 당 골 횟수 꺾은선 그래프

homeTeam = '맨시티'
nameDf = df[df['홈'] == homeTeam]
hteamScore = nameDf['득점']
time = nameDf['시간']

awayTeam = '리버풀'
name = df[df['원정'] == awayTeam]
ateamScore= name['득점']


x = time
y1 = hteamScore
y2 = ateamScore

plt.plot(x, y1)
plt.plot(x, y2)
plt.show()
```

```python
#홈팀과 원정팀의 승리 횟수 비교 분석 막대그래프

homeTeam = '맨시티'
home = df[df['홈'] == homeTeam]

gain = home['득점'].values.tolist()
loss = home['실점'].values.tolist()

count = 0
for i in range(len(gain)):
    if gain[i] > loss[i]:
        count += 1

        
awayTeam = '리버풀'
away = df[df['원정'] == awayTeam]

gain = home['득점'].values.tolist()
loss = home['실점'].values.tolist()

count2 = 0
for i in range(len(gain)):
    if gain[i] > loss[i]:
        count2 += 1
        
x1 = homeTeam
x2 = awayTeam

y1 = count
y2 = count2

plt.bar(x1, y1)
plt.bar(x2, y2)
plt.show()
```
